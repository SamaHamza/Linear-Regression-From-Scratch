#  ***Linear Regression***

---




```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
trainingExample = [
(1,1),
(2,3),
(4,3),
]
```

```python
# defining the function or the predictor:
def fun(x):
  return np.array([1,x])
```

```python
def weightvec():
  return np.zeros(2)
  #where the 2 is the dimension of the array wrt the fun(x)
```

```python
# now we need to define MSE as a loss function:
def trainLoss(w):
  return 1.0 / len(trainingExample) * sum((w.dot(fun(x)) - y)**2 for x,y in trainingExample)
```

```python
# now we need to define optimizer function aka gradient for the loss function:
def GradientTrainLoss(w):
  return 1.0 / len(trainingExample) * sum( 2*(w.dot(fun(x)) - y) * fun(x) for x,y in trainingExample)
```

```python
# Gradient descent Function:
def GradientDescentFun(F, gradientF, weightvec):
  w = weightvec()
  iter = 0.1
  mse_list = []

  for t in range(50): # 50 is the number of epochs
    value = F(w)
    gradient = gradientF(w)
    w = w - iter * gradient

    # Calculate and store the MSE for the current epoch
    mse = F(w)
    mse_list.append(mse)



    print(f'epoch {t}: w ={w}, F(w)= value, gradientF ={gradient}, , MSE = {mse}')
    # Calculate and print the average MSE
    avg_mse = sum(mse_list) / len(mse_list)
    print(f'Average MSE: {avg_mse}')
    print("")

    print(f'MSE list: {mse_list}')

    # Plotting the MSE values
    plt.plot(range(len(mse_list)), mse_list)
    plt.xlabel('Epoch')
    plt.ylabel('Mean Squared Error (MSE)')
    plt.title('MSE vs Epoch')
    plt.show()
```

```python
GradientDescentFun(trainLoss, GradientTrainLoss, weightvec)
```

```python
# Plotting the linear regression line after the calling is over
w = weightvec()
iter = 0.1
for t in range(500):
    value = trainLoss(w)
    gradient = GradientTrainLoss(w)
    w = w - iter * gradient

plt.scatter([x for x, _ in trainingExample], [y for _, y in trainingExample], label="Training Data")
plt.plot([x for x, _ in trainingExample], [w.dot(fun(x)) for x, _ in trainingExample], color='red', label="Regression Line")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
